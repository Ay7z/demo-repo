```python
import pandas as pd

data = {
    "Student": ["Rahul", "Rahul", "Rahul", "Rahul",
                "Priya", "Priya", "Priya", "Priya",
                "Amit",  "Amit",  "Amit",  "Amit"],
    "Subject": ["Maths", "Science", "English", "Hindi",
                "Maths", "Science", "English", "Hindi",
                "Maths", "Science", "English", "Hindi"],
    "Sem":     [1, 1, 2, 2, 1, 1, 2, 2,
                1, 1, 2, 2],
    "Marks":   [85, 72, 90, 78,
                91, 88, 76, 82,
                78, 65, 82, 70]
}

df = pd.DataFrame(data)
```

```python
pd.pivot_table(df, index ='Student', columns = 'Subject', values = 'Marks', aggfunc = 'sum')
```

```python
pd.pivot_table(df, index='Sem', columns = 'Student', values = 'Marks' , aggfunc='mean')
```

```python
pd.pivot_table(df, index='Subject', columns = 'Sem', values = 'Marks', aggfunc='max')
```

```python
print(df[df["Subject"] != "Hindi"])
```

```python
pd.pivot_table(df, index = 'Student', columns = 'Sem', values = 'Marks', aggfunc = 'count')
```

```python
import pandas as pd

data = {
    "Student": ["Rahul", "Rahul", "Rahul",
                "Priya", "Priya", "Priya",
                "Amit",  "Amit",  "Amit"],
    "Subject": ["Maths", "Science", "English",
                "Maths", "Science", "English",
                "Maths", "Science", "English"],
    "Sem":     [1, 1, 2, 1, 2, 2, 1, 2, 1],
    "Marks":   [85, 72, 90, 91, 88, 76, 78, 65, 82]
}

df = pd.DataFrame(data)
```

```python
df.groupby('Student')['Marks'].min()
```
