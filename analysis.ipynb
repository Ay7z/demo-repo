```python
import pandas as pd
import numpy as np

data = {
    "Employee": ["Rahul", "Priya", "Amit", "Sneha", "Ravi",
                 "Neha", "Vijay", "Pooja", "Arjun", "Divya"],
    "Dept":     ["Sales", "Tech", "Sales", "HR", "Tech",
                 "HR", "Sales", "Tech", "Sales", "HR"],
    "City":     ["Hyderabad", "Mumbai", "Delhi", "Hyderabad", "Mumbai",
                 "Delhi", "Hyderabad", "Mumbai", "Delhi", "Hyderabad"],
    "Salary":   [45000, 72000, 51000, 38000, 85000,
                 42000, 55000, 68000, 49000, 500000],
    "Experience":[2, 5, 3, 1, 7, 2, 4, 6, 3, 2]
}

df = pd.DataFrame(data)
```

```python
df[(df['Salary'] > 40000) & (df['Salary'] < 70000)],['Employees']
```

```python

df[(df['City'] == "Mumbai") | (df['City'] == 'Delhi') & (df['Experience'] > 4)]
```

```python
median_Salary = df['Salary'].median()
df[df['Salary'] > median_Salary]
```

```python
df.sort_values('Experience', ascending=False).groupby('Dept').head(3)
```

```python
df[(df['City'] == "Hyderabad") & (df['Salary'] > 50000)]
```

```python
Q1 =df['Salary'].quantile(0.25)
Q3 =df['Salary'].quantile(0.75)
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 2 * IQR
upper_limit = Q3 + 2 * IQR
```

```python
lower_limit
```

```python
df[(df["Salary"] >= lower_limit) & (df['Salary'] <= upper_limit)]
```
